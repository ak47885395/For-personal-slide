```python
within_two_hour = 2*60*60
time_back_two_hour = time.localtime(time.time()-within_two_hour)

print(time.strftime('%Y %m %d %H:%M:%S'  , time_back_two_hour  ))


browser = webdriver.Chrome("chromedriver.exe")

browser.get("https://www.ettoday.net/news/news-list.htm")


while(1):
    
            html = browser.page_source
        
            soup = BeautifulSoup(html)
            
            print("sucess")
            
            last = soup.find(class_="part_list_2").find_all("h3")[-1]
            
            last_time =  time.strptime(  last.find(class_="date").text  ,  "%Y/%m/%d %H:%M")
            
                
            
            time.sleep(2)
            
            browser.execute_script("window.scrollTo( 0 ,  document.body.scrollHeight  )")
            
            if (last_time.tm_hour <= time_back_two_hour.tm_hour): #& (element_time.tm_min < time_back_two_hour.tm_min) :
                
                break     #####
                              #####   無法確認 滑下去 最後一個 小時是多少  只能 <= 就結束
                 ############################
            
for  i  in soup.find(class_="part_list_2").find_all("h3"):
    
    #  逐項比對時間
    
    element_time = time.strptime( i.find(class_="date").text ,  "%Y/%m/%d %H:%M" )
    
  #  if elemnt_time.tm_sec < time_back_two_hour.tm_sec:
    
    
 #   if element_time.tm_hour != time_back_two_hour.tm_hour :
          
    if  (element_time.tm_hour >= time_back_two_hour.tm_hour) :
        
        if (element_time.tm_hour == time_back_two_hour.tm_hour) & (element_time.tm_min < time_back_two_hour.tm_min):
        
            break
        
        else:
        
            print(i.text)   
```
